# plot
> Plotting functionality

```python
#| export
import torch
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#| export
sns.set_theme()
```

A function to make a *scatterplot*, along with some marks along the horizontal axis.

```python
#| export
def partially_observed_time_series(
    observed_t: torch.Tensor,
    observed_y: torch.Tensor,
    to_predict_at_t: torch.Tensor
):
    
    _, axis = plt.subplots()
    
    sns.scatterplot(x=observed_t, y=observed_y, color='blue')
    
    for x in to_predict_at_t:
        axis.axvline(x, color='gray', linestyle='dashed')
        
    axis.set(xlabel='t', ylabel='observations')
    
    return axis
```

Some test data

```python
t = torch.distributions.uniform.Uniform(0., 5.).sample((10,))
t = torch.sort(t)[0]
y = torch.randn_like(t)
```

```python
partially_observed_time_series(t[:8], y[:8], t[8:]);
```
